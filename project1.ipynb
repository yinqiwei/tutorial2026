```python
import math
from itertools import combinations

# -------------------- Input Example Data (in bohr) --------------------
DATA = """\
7
6 0.000000000000 0.000000000000 0.000000000000
6 0.000000000000 0.000000000000 2.845112131228
8 1.899115961744 0.000000000000 4.139062527233
1 -1.894048308506 0.000000000000 3.747688672216
1  1.942508819960 0.000000000000 -0.701145981971
1 -1.007295466862 -1.669971842687 -0.705916966833
1 -1.007295466862  1.669971842687 -0.705916966833
"""

# Element symbols and atomic masses (amu)
Z2SYM = {1: "H", 6: "C", 8: "O"}
Z2M = {1: 1.00784, 6: 12.0106, 8: 15.999}

# -------------------- Physical Constants --------------------
BOHR_TO_ANG = 0.529177210903
AMU_TO_G = 1.66053906660e-24
ANG_TO_CM = 1.0e-8
H_CGS = 6.62607015e-27        # erg*s
C_CMS = 2.99792458e10         # cm/s

# -------------------- Linear Algebra -------------------------
def sub(a, b):
    return (a[0]-b[0], a[1]-b[1], a[2]-b[2])

def dot(a, b):
    return a[0]*b[0] + a[1]*b[1] + a[2]*b[2]

def cross(a, b):
    return (
        a[1]*b[2] - a[2]*b[1],
        a[2]*b[0] - a[0]*b[2],
        a[0]*b[1] - a[1]*b[0],
    )

def norm(v):
    return math.sqrt(dot(v, v))

def safe_unit(vec, eps=1e-14):
    n = norm(vec)
    if n < eps:
        return None
    return (vec[0]/n, vec[1]/n, vec[2]/n)

def unit_vec(R, a, b):
    """Unit vector: a → b"""
    return safe_unit(sub(R[b], R[a]))

# -------------------- Parsing Input --------------------
def parse_block(text: str):
    lines = [ln.strip() for ln in text.splitlines() if ln.strip()]
    N = int(lines[0].split()[0])
    Z, R = [], []
    for k in range(N):
        toks = lines[k + 1].split()
        znum = int(float(toks[0]))
        x, y, zc = float(toks[1]), float(toks[2]), float(toks[3])
        Z.append(znum)
        R.append((x, y, zc))
    return Z, R

# -------------------- Distances and Bond Angles --------------------
def bond_length(R, i: int, j: int) -> float:
    return norm(sub(R[i], R[j]))

def angle_ijk(R, i, j, k):
    """∠i-j-k at atom j -> (cosφ, φ[rad], φ[deg])"""
    u1 = unit_vec(R, j, i)
    u2 = unit_vec(R, j, k)
    if u1 is None or u2 is None:
        return None
    c = max(-1.0, min(1.0, dot(u1, u2)))
    phi = math.acos(c)
    return c, phi, math.degrees(phi)

# -------------------- Out-of-Plane Angles --------------------
def out_of_plane_i_jkl(R, i, j, k, l):
    """θ_{i|jkl}: atom i out of the plane defined by j-k-l (k is the central atom)"""
    e_kj = unit_vec(R, k, j)
    e_kl = unit_vec(R, k, l)
    e_ki = unit_vec(R, k, i)
    if e_kj is None or e_kl is None or e_ki is None:
        return None
    n = cross(e_kj, e_kl)
    den = norm(n)
    if den < 1e-14:
        return None
    s = dot(n, e_ki) / den
    s = max(-1.0, min(1.0, s))
    th = math.asin(s)
    return s, th, math.degrees(th)

# -------------------- Torsion (Dihedral) Angles --------------------
def torsion_ijkl(R, i, j, k, l):
    """Signed dihedral angle τ_{ijkl}: returns (cosτ, sinτ, τ[rad], τ[deg])"""
    e_ij = unit_vec(R, i, j)
    e_jk = unit_vec(R, j, k)
    e_kl = unit_vec(R, k, l)
    if e_ij is None or e_jk is None or e_kl is None:
        return None
    n1 = cross(e_ij, e_jk)
    n2 = cross(e_jk, e_kl)
    n1n, n2n = norm(n1), norm(n2)
    if n1n < 1e-14 or n2n < 1e-14:
        return None
    n1u = (n1[0]/n1n, n1[1]/n1n, n1[2]/n1n)
    n2u = (n2[0]/n2n, n2[1]/n2n, n2[2]/n2n)
    x = max(-1.0, min(1.0, dot(n1u, n2u)))
    y = dot(e_jk, cross(n1u, n2u))
    tau = math.atan2(y, x)
    return x, y, tau, math.degrees(tau)

# -------------------- Center of Mass --------------------
def masses_from_Z(Z):
    return [Z2M[z] for z in Z]

def center_of_mass(Z, R):
    m = masses_from_Z(Z)
    mtot = sum(m)
    xcm = sum(mi * Ri[0] for mi, Ri in zip(m, R)) / mtot
    ycm = sum(mi * Ri[1] for mi, Ri in zip(m, R)) / mtot
    zcm = sum(mi * Ri[2] for mi, Ri in zip(m, R)) / mtot
    return (xcm, ycm, zcm), mtot

def translate_to_com(R, com):
    dx, dy, dz = com
    return [(x - dx, y - dy, z - dz) for (x, y, z) in R]

# -------------------- Inertia Tensor --------------------
def inertia_tensor_about_COM(Z, R):
    (xcm, ycm, zcm), _ = center_of_mass(Z, R)
    Rc = translate_to_com(R, (xcm, ycm, zcm))
    m = masses_from_Z(Z)

    Ixx = Iyy = Izz = Ixy = Ixz = Iyz = 0.0
    for mi, (x, y, z) in zip(m, Rc):
        Ixx += mi * (y*y + z*z)
        Iyy += mi * (x*x + z*z)
        Izz += mi * (x*x + y*y)
        Ixy += mi * (x*y)
        Ixz += mi * (x*z)
        Iyz += mi * (y*z)

    I = [
        [ Ixx, -Ixy, -Ixz ],
        [ -Ixy, Iyy, -Iyz ],
        [ -Ixz, -Iyz, Izz ],
    ]

    import numpy as np
    vals, vecs = np.linalg.eigh(np.array(I, dtype=float))
    vals = sorted(vals.tolist())
    return I, vals

def convert_I_units(I_amu_bohr2):
    bohr2_to_ang2 = BOHR_TO_ANG**2
    I_amu_ang2 = [x * bohr2_to_ang2 for x in I_amu_bohr2]
    I_g_cm2 = [x * AMU_TO_G * (ANG_TO_CM**2) for x in I_amu_ang2]
    return I_amu_ang2, I_g_cm2

# --------------------  Rotational Constants and Classification --------------------
def rotational_constants_from_I_gcm2(Ia, Ib, Ic):
    denom_cm = 8.0 * math.pi**2 * C_CMS
    denom_hz = 8.0 * math.pi**2
    A_cm = H_CGS / (denom_cm * Ia)
    B_cm = H_CGS / (denom_cm * Ib)
    C_cm = H_CGS / (denom_cm * Ic)
    A_Hz = H_CGS / (denom_hz * Ia)
    B_Hz = H_CGS / (denom_hz * Ib)
    C_Hz = H_CGS / (denom_hz * Ic)
    return (A_cm, B_cm, C_cm), (A_Hz/1e6, B_Hz/1e6, C_Hz/1e6)

def classify_rotor(N, Ia, Ib, Ic, tol_linear=1e-6, tol_equal=1e-3):
    if N == 2:
        return "diatomic"
    if Ia < tol_linear * Ic:
        return "linear polyatomic"
    def rel(a,b): return abs(a-b)/max(a,b)
    if rel(Ia,Ib) < tol_equal and rel(Ib,Ic) < tol_equal:
        return "spherical top"
    if rel(Ib,Ic) < tol_equal and Ia < Ib:
        return "symmetric top (prolate)"
    if rel(Ia,Ib) < tol_equal and Ic > Ib:
        return "symmetric top (oblate)"
    return "asymmetric top"

# -------------------- Interactive Main Program --------------------
def main():
    Z, R = parse_block(DATA)
    R_orig = R[:]
    N = len(Z)

    print("Coordinates loaded (unit: bohr)")
    print("Idx  Z  El    x                    y                    z")
    print("-" * 74)
    for idx, (znum, (x, y, zc)) in enumerate(zip(Z, R), start=1):
        sym = Z2SYM.get(znum, "?")
        print(f"{idx:>3}  {znum:2d}  {sym:>2}  {x: .12f}  {y: .12f}  {zc: .12f}")

    print("\nCommands:")
    print("  d i j          -> Bond length R(i,j)")
    print("  a i j k        -> Bond angle ∠i-j-k (center j)")
    print("  list j         -> List all bond angles around j")
    print("  oop i j k l    -> Out-of-plane angle θ_{i|jkl} (k as central)")
    print("  tors i j k l   -> Torsional (dihedral) angle τ_{ijkl}")
    print("  com            -> Compute center of mass (does not modify coords)")
    print("  center         -> Preview translation to COM")
    print("  center!        -> Translate to COM (update R)")
    print("  reset          -> Restore original coordinates")
    print("  rotor          -> Inertia tensor → principal moments → rotational constants → type")
    print("  q              -> Quit")

    while True:
        s = input("\nCommand: ").strip().lower()
        if s in {"q", "quit", "exit"}:
            break
        if not s:
            continue
        toks = s.split()
        cmd = toks[0]

        # Bond length
        if cmd == "d" and len(toks) == 3 and all(t.isdigit() for t in toks[1:]):
            i, j = int(toks[1])-1, int(toks[2])-1
            if 0 <= i < N and 0 <= j < N and i != j:
                print(f"R({i+1},{j+1}) = {bond_length(R, i, j):.10f} bohr")
            continue

        # Bond angle
        if cmd == "a" and len(toks) == 4 and all(t.isdigit() for t in toks[1:]):
            i, j, k = int(toks[1])-1, int(toks[2])-1, int(toks[3])-1
            res = angle_ijk(R, i, j, k)
            if res:
                c, rad, deg = res
                print(f"∠({i+1}-{j+1}-{k+1}): cosφ={c:.12f}, φ={rad:.10f} rad = {deg:.6f}°")
            continue

        # List all angles around j
        if cmd == "list" and len(toks) == 2 and toks[1].isdigit():
            j = int(toks[1]) - 1
            others = [idx for idx in range(N) if idx != j]
            for i, k in combinations(others, 2):
                res = angle_ijk(R, i, j, k)
                if res:
                    _, _, deg = res
                    print(f"∠({i+1}-{j+1}-{k+1}) = {deg:10.6f}°")
            continue

        # Out-of-plane
        if cmd == "oop" and len(toks) == 5 and all(t.isdigit() for t in toks[1:]):
            i, j, k, l = (int(toks[1])-1, int(toks[2])-1, int(toks[3])-1, int(toks[4])-1)
            res = out_of_plane_i_jkl(R, i, j, k, l)
            if res:
                s_val, rad, deg = res
                print(f"θ_{{{i+1}|{j+1}-{k+1}-{l+1}}}: sinθ={s_val:.12f}, θ={rad:.10f} rad = {deg:.6f}°")
            continue

        # Torsion
        if cmd == "tors" and len(toks) == 5 and all(t.isdigit() for t in toks[1:]):
            i, j, k, l = (int(toks[1])-1, int(toks[2])-1, int(toks[3])-1, int(toks[4])-1)
            res = torsion_ijkl(R, i, j, k, l)
            if res:
                cosv, sinv, rad, deg = res
                print(f"τ_{i+1}-{j+1}-{k+1}-{l+1}: cosτ={cosv:.12f}, sinτ={sinv:.12f}, τ={rad:.10f} rad = {deg:.6f}°")
            continue

        # COM
        if cmd == "com" and len(toks) == 1:
            com, mtot = center_of_mass(Z, R)
            print(f"Center of Mass (bohr): X={com[0]:.12f}  Y={com[1]:.12f}  Z={com[2]:.12f}  Total Mass={mtot:.6f} amu")
            print(f"(Å): X={com[0]*BOHR_TO_ANG:.12f}  Y={com[1]*BOHR_TO_ANG:.12f}  Z={com[2]*BOHR_TO_ANG:.12f}")
            continue

        # Center preview
        if cmd == "center" and len(toks) == 1:
            com, mtot = center_of_mass(Z, R)
            shift = (-com[0], -com[1], -com[2])
            print(f"[Preview] COM (bohr): {com}")
            print(f"[Preview] Translation vector: {shift}")
            print("Type 'center!' to actually apply translation.")
            continue

        # Apply COM translation
        if cmd == "center!" and len(toks) == 1:
            com, _ = center_of_mass(Z, R)
            R[:] = translate_to_com(R, com)
            print("Coordinates have been translated to the center of mass.")
            continue

        # Reset to original
        if cmd == "reset" and len(toks) == 1:
            R[:] = R_orig[:]
            print("Coordinates restored to original.")
            continue

        # Rotor analysis
        if cmd == "rotor" and len(toks) == 1:
            I_mat, vals_amu_bohr2 = inertia_tensor_about_COM(Z, R)
            Ia, Ib, Ic = vals_amu_bohr2
            I_amu_ang2, I_g_cm2 = convert_I_units(vals_amu_bohr2)
            Ia_g, Ib_g, Ic_g = I_g_cm2

            print("Inertia tensor about COM (amu·bohr^2):")
            for row in I_mat:
                print("   " + "  ".join(f"{v: .10f}" for v in row))
            print(f"Principal moments (amu·bohr^2): Ia={Ia:.10f}, Ib={Ib:.10f}, Ic={Ic:.10f}")
            print(f"(amu·Å^2): Ia={I_amu_ang2[0]:.10f}, Ib={I_amu_ang2[1]:.10f}, Ic={I_amu_ang2[2]:.10f}")
            print(f"(g·cm^2): Ia={Ia_g:.6e}, Ib={Ib_g:.6e}, Ic={Ic_g:.6e}")

            (A_cm, B_cm, C_cm), (A_MHz, B_MHz, C_MHz) = rotational_constants_from_I_gcm2(
                Ia_g, Ib_g, Ic_g
            )
            print("Rotational constants:")
            print(f"  A = {A_cm:.6f} cm^-1  ({A_MHz:.3f} MHz)")
            print(f"  B = {B_cm:.6f} cm^-1  ({B_MHz:.3f} MHz)")
            print(f"  C = {C_cm:.6f} cm^-1  ({C_MHz:.3f} MHz)")

            mtype = classify_rotor(N, Ia_g, Ib_g, Ic_g)
            print(f"Molecular type: {mtype}")
            continue

        # Fallback for unknown commands
        print(
            "Invalid command. Examples: d 1 3 | a 1 2 3 | list 2 | "
            "oop 1 2 3 4 | tors 1 2 3 4 | com | center | center! | reset | rotor | q"
        )

if __name__ == "__main__":
    main()
```
